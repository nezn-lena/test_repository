```python
# Метод Z-оценка. измеряет сколько стандартных отклонений точка данных отстоит от среднего значения.
# m - среднее значение признака .mean()
# s - отклонение признака .std()
# Построить интервал [m-3s,m+3s]. Все объекты, значения признака которых не попадет в этот интервал, считаем выбросами. Их удаляем из таблицы.

m = df['Weight'].mean()
s = df['Weight'].std()
m-3*s, m+3*s
```

```python
df = df[ (df['Weight']<m+3*s) & (df['Weight']>m-3*s) ]
sns.boxplot(x = df['Weight'])
```

```python
# Метод IQR (межквартильный размах)
# a - 25-ая пронцентиль признака
# b - 75-ая пронцентиль признака
# Построить интервал [a-1,5(b-a),a+1,5(b-a)]
# Все объекты, значения признака которых не попадет в этот интервал, считаем выбросами. Их удаляем из таблицы.

df['Age'].quantile(0.25), df['Age'].quantile(0.75)
```

```python
a = df['Age'].quantile(0.25)
b = df['Age'].quantile(0.75)
df = df[ ( df['Age']<b+1.5*(b-a) ) & ( df['Age']>a-1.5*(b-a) ) ]

sns.boxplot(x = df['Age'])
```

```python
# линейная регрессия
# предсказание числового признака - это задача регрессии

import pandas as pd
import seaborn as sns
from sklearn.linear_model import LinearRegression

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/students.csv', delimiter=',')

df_cut = df[['Growth','Shoe size']]
df_cut = df_cut.dropna() # удалили все пропуски в таблице
```

```python
# Гипотеза - рост линейно зависит от размера обуви, т.е.
# 'Growth' = a * 'Shoe size' + b
# или 'Growth' = result.coef_ * 'Shoe size' + result.intercept_
LinearRegression = LinearRegression()
result = LinearRegression.fit(df_cut['Shoe size'].values.reshape(-1,1), y=df_cut['Growth'].values)
# 1 параметр не целевой, по которому идет предсказание. 2й параметр (y) целевой признак, который предсказываем.
# х - преобразуем в вертикальный массив чисел, y - в горизонтальный массив чисел

result.coef_, result.intercept_
```

```python
df_cut['Predict Growth'] = result.predict(df_cut['Shoe size'].values.reshape(-1,1)) # добавляем предсказанные значения
```

# Алгоритмы классификации. Метод k ближайших соседей

```python
# Гипотеза: мужчины и женщины образуют два облака. Причем для многих точек выполнено условие:
# если соседние точки мужчины (женщины), то и данная точка - мужчина(женщина)
import pandas as pd
import seaborn as sns
from sklearn.neighbors import KNeighborsClassifier

df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/students.csv', delimiter=',')

df_cut = df[['Weight','Growth', 'Sex']]
df_cut = df_cut.dropna() # удалили все пропуски в таблице
```

```python
sns.scatterplot(data = df_cut, x='Weight', y='Growth', hue='Sex') # визуализируем для наглядности
```

```python
# необходимо нормировать разнородные показатели - вес в кг, а рост в см.
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
scaler.fit(df_cut[['Weight','Growth']].values.reshape(-1,2))
arr=scaler.transform(df_cut[['Weight','Growth']].values.reshape(-1,2))
```

```python
model = KNeighborsClassifier(n_neighbors=1) # параметр в скобках фиксирует сколько ближайших соседей будем рассматривать
model.fit(arr, y=df_cut['Sex'].values)
```

```python
#загружаем тестовую выборку
df_test = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/students_test.csv', delimiter=',')
df_test_cut = df_test[['Weight','Growth', 'Sex']]
df_test_cut = df_test_cut.dropna() # удалили все пропуски в таблице
```

```python
arr_test = scaler.transform(df_test_cut[['Weight','Growth']].values.reshape(-1,2)) #нормируем тестовую выборку
```

```python
df_test_cut['Predicted'] = model.predict(arr_test)
```

Метрики качества в задаче классификации


---


recall полнота.  показывает, какую долю положительных объектов модель смогла правильно идентифицировать.\
recall для женщин (для мужчин) = доля правильно предсказанных женщин(мужчин) среди всех женщин(мужчин)\
 40/(40+2)=0.97560976


---


precision точность. показывает, какая доля объектов, выделенных моделью как положительные, действительно являются положительными.\
precision  для женщин (для мужчин) = доля истинных женщин(мужчин) среди людей, отнесенных моделью к женщинам(мужчинам)\
 40/(40+1)=0.95238095
