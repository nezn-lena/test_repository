# Series (Серия)
— это одномерный массив данных, у которого есть метки (индекс).

```python
# Создание Series из списка (List)
data_list = [10, 20, 30, 40]

series_from_list = pd.Series(data_list) # Превращаем в Series
print(series_from_list)
```

```python
# Создание Series с собственным индексом
sales = [150, 200, 100] # Данные - продажи
months = ['Jan', 'Feb', 'Mar'] # Индекс - месяцы

s = pd.Series(sales, index=months) # Явное указание индекса
print(s)
```

```python
# Создание Series из словаря (Dictionary)
currencies = {
    'USD': 92.5,
    'EUR': 99.1,
    'CNY': 12.8
}

s_curr = pd.Series(currencies) # Pandas автоматически поймет, что ключи — это индексы

print(s_curr)
```

```python
# Параметр name
#У Series есть полезный атрибут name. Это имя всей серии (будущее название колонки, если мы объединим несколько серий в таблицу).

s = pd.Series([10, 20, 30], name='Temperature')
print(s)
```

```python
#Записи ниже обращаются к одному и тому же элементу
s_curr[1] # Обратиться по неявной позиции
s_curr['EUR'] # Обратиться по явной метке
```

**!!!** Если явный индекс состоит из целых чисел, pandas при использовании квадратных скобок [] ищет ***только явную метку***. Как избежать путаницы? (loc и iloc)\
**.iloc[]** (Integer LOCation) — всегда смотрит на неявный позиционный индекс (0, 1, 2...).
`s_years.iloc[0] -> вернет 500 (первый элемент).`\
**.loc[]** (LOCation) — всегда смотрит на явный индекс (метку).
`s_years.loc[2010] -> вернет 500.`

# DataFrame (Датафрейм)
— это двумерная табличная структура данных.\
Технически DataFrame — это набор объектов Series, у которых общий индекс.\
У нас есть две оси: 0 (строки) и 1 (столбцы).

```python
# создание DataFrame из словаря списков (Dict of Lists).
# При этом важно чтобы все списки значений словаря должны быть одинаковой длины.
data = {
    'Name': ['Anna', 'Bob', 'Charlie'],
    'Age': [25, 30, 35],
    'City': ['Moscow', 'London', 'Paris']
}

df = pd.DataFrame(data)
print(df)
```

```python
# создание DataFrame из списка словарей (List of Dicts). JSON-стиль
# Каждый словарь — это одна строка. Ключи словарей становятся колонками.
data = [
    {'Name': 'Anna', 'Age': 25, 'City': 'Moscow'},
    {'Name': 'Bob', 'Age': 30, 'City': 'London'},
    {'Name': 'Charlie', 'Age': 35}  # У Чарли нет города!
]

df = pd.DataFrame(data)
print(df)
```

```python
# создание DataFrame из писка списков (List of Lists) или Массив NumPy
# Имена колонок нужно передать отдельно через параметр columns.
raw_data = [
    ['Anna', 25],
    ['Bob', 30],
    ['Charlie', 35]
]

df = pd.DataFrame(raw_data, columns=['Name', 'Age'])
print(df)
```

```python
# создание DataFrame из массива NumPy
# Это стандарт для Machine Learning. Если у вас есть матрица чисел (например, картинка или признаки для нейросети), её легко превратить в DataFrame.

import numpy as np

# Создаем матрицу 3x2 (3 строки, 2 столбца)
arr = np.array([
    [10, 20],
    [30, 40],
    [50, 60]
])

df = pd.DataFrame(arr, columns=['Feature_1', 'Feature_2'])
print(df)
```

## `loc[] `
Атрибут .loc (от слова Location) используется для доступа к данным по их МЕТКАМ (Labels).
Порядок: [строка, столбец]\

```python
from operator import index
import pandas as pd

data = {
    'Country': ['Russia', 'France', 'Italy', 'Germany'],
    'Capital': ['Moscow', 'Paris', 'Rome', 'Berlin'],
    'Population': [144, 67, 60, 83]
}

# Делаем страну индексом!
df = pd.DataFrame(data, index=data['Country'])
df = df[['Capital','Population']]
# (теперь строки называются 'Russia', 'France'...)
```

```python
# Если мы передадим в .loc[] только один аргумент, pandas поймет, что мы ищем строку.
df.loc[['France','Russia']]
```

```python
# получить значение в конкретной ячейке или Можно брать подмножества (срез таблицы):

df.loc[['Russia', 'Germany'], 'Capital']
```

```python
df.loc['France':'Germany'] # Срезы в .loc ВКЛЮЧАЮТ оба конца (Inclusive).
```

```python
df.loc[df['Population'] > 80] # В .loc можно передавать не только конкретные имена, но и условия
```

## `.iloc[] `
Атрибут .iloc[] (от слова Integer Location) используется для доступа к данным исключительно по их численной позиции.\
df.iloc[номер_строки, номер_столбца]

```python
df.iloc[0:3,0:2] # В .iloc срезы работают как в обычных списках Python: Старт — включается. Конец (Stop) — НЕ включается (Exclusive).
```

```python
df.iloc[[0, 3], [1]] # можете передать конкретный набор номеров, которые вам нужны.
```

```python
# Отрицательная индексация
df.iloc[-1] # Последняя строка таблицы
df.iloc[:, -1] # Все строки, но только последний столбец
```

## Отбор строк по условию

```python
df[(df['Population'] > 80) & (df['Capital']=='Moscow')] # отобрать строки по условию

# Логика Python Pandas Описание
# И (Оба условия верны) and & Амперсанд
# ИЛИ (Хотя бы одно верно) or | Разделитель
# НЕ (Инверсия) not ~ Тильда\
```

```python
df[df['Capital'].isin(['Moscow','Rome'])] # Метод .isin() Множественный выбор, аналог SQL-оператора IN (...).
```

```python
# А если нужно выбрать всех, кто НЕ живет в этих городах? Просто поставьте ~ перед условием.
df[~df['Capital'].isin(['Moscow','Rome'])]
```

```python
# Метод .between() (Диапазон) проверяет попадание в интервал.
df[df['Population'].between(60, 100)]
# По умолчанию .between включает границы (inclusive).
# Если нужно строгое неравенство, используйте параметр inclusive='neither'
```

```python
# Метод .query() (SQL-стиль). Метод позволяет писать условия одной строкой внутри кавычек.

df.query("Population > 80 and Capital=='Moscow'")
```

```python
# Самый крутой трюк .query() — это возможность ссылаться на внешние переменные Python.
# Для этого перед именем переменной ставится собачка @.

min_pop = 18
target_city = 'Moscow'

# Pandas поймет, что @min_age — это переменная из вашего кода, а Age — колонка
df.query("Population > @min_pop and Capital == @target_city")
```

```python
# Изменение типов данных: метод .astype()
df = pd.DataFrame({
    'user_id': [1, 2, 3],
    'price': ['100', '250', '500'], # Это строки!
    'is_active': [0, 1, 1]          # Это целые числа (0 и 1)
})

print(df.dtypes)
# price        object
# is_active     int64

# Перезаписываем колонку 'price' её же измененной версией
df['price'] = df['price'].astype(int)

# То же самое с 'is_active'
df['is_active'] = df['is_active'].astype(bool)

print(df.dtypes)
# price        int32 (или int64)
# is_active     bool
```

## Обнаружение пропусков: isna() и isnull()

```python
import pandas as pd
import numpy as np

# Создаем таблицу с дырками (np.nan - это специальный объект "пустоты")
df = pd.DataFrame({
    'Name': ['Anna', 'Bob', 'Charlie'],
    'Age': [25, np.nan, 35],       # У Боба нет возраста
    'City': ['Moscow', 'Paris', None] # У Чарли нет города
})

print(df.isna()) # вернет нам таблицу точно такого же размера, но состоящую только из True и False.
# True — значение пропущено (ячейка пустая).
# False — значение есть (всё в порядке).
```

```python
# никогда не используют df.isna() в чистом виде. Его всегда комбинируют с методом подсчета суммы — .sum().
# В Python логическое True равно 1, а False равно 0. Если мы просуммируем маску по столбцам, то получим точное количество пропусков в каждой колонке!
print(df.isna().sum())
```

## Множественная агрегация с методом .agg()

```python
# Метод .agg() позволяет применить сразу несколько математических функций за один проход.
import pandas as pd

df = pd.DataFrame({
    'Department': ['IT', 'IT', 'Sales', 'Sales', 'HR'],
    'Name': ['Anna', 'Bob', 'Charlie', 'David', 'Eve'],
    'Salary': [120000, 80000, 90000, 110000, 70000],
    'Age': [25, 30, 35, 40, 28]
})

# Смотрим статистику зарплат по каждому отделу
salary_stats = df.groupby('Department')['Salary'].agg(['min', 'max', 'mean'])

print(salary_stats)
```

```python
# Разные функции к разным столбцам.
# в .agg() нужно передать словарь dict.
# Ключи словаря — это названия колонок.
# Значения — это функции (или списки функций), которые к ним нужно применить.

# Для возраста считаем среднее, для зарплаты - сумму, для имени - количество
complex_stats = df.groupby('Department').agg({
    'Age': 'mean',
    'Salary': 'sum',
    'Name': 'count'
})

print(complex_stats)
```

```python
# Идеальная сводная таблица с понятными именами
perfect_stats = df.groupby('Department', as_index=False).agg(
    avg_age=('Age', 'mean'),           # Новая колонка avg_age
    total_budget=('Salary', 'sum'),    # Новая колонка total_budget
    employee_count=('Name', 'count')   # Новая колонка employee_count
)

print(perfect_stats)
```

```python
# Настраиваемые функции агрегации (lambda)

df.groupby('Department')['Salary'].agg(lambda x: x.max() - x.min())
```

```python
# посчитаем среднюю зарплату по отделу, а также соберем имена всех сотрудников этого отдела в одну строку через запятую.
# Для склеивания строк в Python используется метод ', '.join(список). Применим его через лямбду:

report = df.groupby('Department', as_index=False).agg(
    avg_salary=('Salary', 'mean'),
    # x - это Series с именами. Мы склеиваем их через запятую
    employees_list=('Name', lambda x: ', '.join(x))
)

print(report)
```

```python
# Например, мы хотим узнать, сколько сотрудников в каждом отделе получают зарплату строго больше 100 000.
rich_count = df.groupby('Department')['Salary'].agg(
    lambda x: (x > 100000).sum()
)

print(rich_count)
```

## Метод transform() для вещания результатов на весь DataFrame

## Изменение регистра и обрезка пробелов

```python
# Главное правило при работе с текстом в pandas:
# чтобы применить строковый метод к целой колонке, перед названием метода всегда нужно писать слово .str (от слова string — строка).

import pandas as pd

df = pd.DataFrame({
    'Name': [' aNNa ', 'BOB', 'charlie  '],
    'City': ['Moscow', ' moscow ', 'MOSCOW']
})

print(df)
```

```python
df['Name_clean'] = df['Name'].str.title() # делает первую букву каждого слова заглавной

df['City_clean'] = df['City'].str.lower() # делает все буквы строчными (маленькими)

df['City_clean'] = df['City'].str.upper() # делает все буквы прописными (БОЛЬШИМИ)

df['Name_clean'] = df['Name'].str.capitalize() # делает заглавной только самую первую букву в строке, остальные опускает.
```

```python
# Обрезка невидимых пробелов (strip)

df['City_clean'] = df['City_clean'].str.strip() # удаляет пробелы с обоих концов строки
df['Name_clean'] = df['Name_clean'].str.lstrip() # удаляет пробелы только слева (Left)
df['Name_clean'] = df['Name_clean'].str.rstrip() # удаляет пробелы только справа (Right)

print(df['City_clean'].tolist())
```

```python
# Метод цепочки (Chaining)
# Каждый новый строковый метод требует своего собственного .str!
# сразу делаем все буквы строчными и удаляем лишние пробелы
df['City_perfect'] = df['City'].str.lower().str.strip()
```

## Поиск подстрок и замена (contains, replace)

```python
import pandas as pd

df = pd.DataFrame({
    'profession': ['Data Scientist', 'data analyst', 'System Administrator', 'Senior Analyst'],
    'salary': ['$1000', '1200$', '1500 USD', '$2000']
})

df
```

## Разделение строк и извлечение данных

```python
# Разделение строки: .str.split() - возвращает колонку, где в каждой ячейке лежит список (list)
import pandas as pd

df = pd.DataFrame({
    'full_name': ['Anna Smith', 'Bob Johnson', 'Charlie Brown'],
    'badge_info': ['ID: 101 - Active', 'ID: 102 - Fired', 'ID: 103 - Active']
})

# Разрезаем по пробелу (' ')
names_split = df['full_name'].str.split(' ')

print(names_split)
```

## Аксессор .dt: извлечение года, месяца, дня недели

```python
import pandas as pd

df = pd.DataFrame({
    'purchase_date': pd.to_datetime(['2023-01-15', '2023-05-20', '2024-12-31'])
})

print(df.dtypes)
```

```python
# С помощью .dt мы можем извлечь любой компонент даты в отдельную колонку:
# .dt.year — год
# .dt.month — месяц (числом от 1 до 12)
# .dt.day — день (числом от 1 до 31)

# Разбираем дату на части
df['year'] = df['purchase_date'].dt.year
df['month'] = df['purchase_date'].dt.month
df['day'] = df['purchase_date'].dt.day

print(df)
```

```python
# .dt.dayofweek (или .dt.weekday) — возвращает порядковый номер дня недели.
# Важно: отсчет начинается с нуля! 0 — это понедельник, 6 — это воскресенье.
df['weekday_num'] = df['purchase_date'].dt.dayofweek
df['weekday_num2'] = df['purchase_date'].dt.weekday

# .dt.day_name() — возвращает текстовое название дня недели по умолчанию на английском (Monday, Tuesday...)
df['weekday_name'] = df['purchase_date'].dt.day_name() # Параметр locale='ru_RU' - сработает только если русская локаль установлена в системе

print(df[['purchase_date', 'weekday_num', 'weekday_num2', 'weekday_name']])
```

```python
# .dt.is_month_end — возвращает True, если это последний день месяца
df['is_end_of_month'] = df['purchase_date'].dt.is_month_end

# .dt.is_leap_year — возвращает True, если год високосный (366 дней).
df['leap_year'] = df['purchase_date'].dt.is_leap_year

# .dt.days_in_month — подскажет, сколько всего дней в месяце этой даты (28, 29, 30 или 31)
df['days_month'] = df['purchase_date'].dt.days_in_month

print(df[['purchase_date', 'is_end_of_month','leap_year','days_month']])
```

## Основы ресемплирования (resample)

```python
# брат-близнец метода .groupby(), но созданный исключительно для времени.
# по умолчанию колонка с датами (формата datetime64) должна находиться в индексе датафрейма!

import pandas as pd

# Генерируем даты за 40 дней (начиная с 1 января 2023 года)
dates = pd.date_range(start='2023-01-01', periods=40, freq='D')

df = pd.DataFrame({
    'date': dates,
    'sales': [10] * 40  # Пусть каждый день продается ровно по 10 штук
})

# Правило №1: Переносим дату в индекс!
df = df.set_index('date')

print(df.head(3))
```

```python
# 'W' (Weekly) — по неделям (конец недели по умолчанию — воскресенье).
# 'M' (Month End) — по месяцам (на последний день месяца).
# 'MS' (Month Start) — по месяцам (на первый день месяца).
# 'Y' или 'A' (Year End) — по годам.
# 'Q' (Quarter) — по кварталам.

# Суммарные продажи по неделям
weekly_sales = df.resample('W')['sales'].sum()
weekly_sales.head(3)
```

```python
# Суммарные продажи по месяцам (на конец месяца)
monthly_sales = df.resample('M')['sales'].sum()
monthly_sales
```

```python
# Ресемплирование без переноса в индекс (Параметр on)

# Возвращаем плоскую таблицу (дата снова стала колонкой)
df_flat = df.reset_index()

# Ресемплируем по кварталам (Q), используя параметр on='date'
# Не забываем про as_index=False, как в groupby, чтобы таблица осталась плоской!
quarter_sales = df_flat.resample('Q', on='date').sum()

# Для плоских таблиц после resample нужно аккуратно сбросить индекс
print(quarter_sales.reset_index(drop=True))
```

# Создание базовых сводных таблиц через pivot()

```python
import pandas as pd

df = pd.DataFrame({
    'date': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02'],
    'city': ['Moscow', 'Kazan', 'Moscow', 'Kazan'],
    'temp': [-5, -10, -2, -8]
})

print(df)
```

```python
# Не работает с дубликатами, он умеет только переставлять ячейки местами. Он не умеет вычислять и агрегировать данные!

# index — какая колонка станет строками (пойдет сверху вниз).
# columns — какая колонка станет новыми заголовками столбцов (пойдет слева направо).
# values — значения из какой колонки заполнят ячейки на пересечении.

df_pivot = df.pivot(index='date', columns='city', values='temp')

print(df_pivot)
```

# Мощный метод pivot_table() (аналог сводных таблиц Excel)

```python
import pandas as pd

df = pd.DataFrame({
    'date': ['2023-01-01', '2023-01-01', '2023-01-01', '2023-01-02'],
    'city': ['Moscow', 'Moscow', 'Kazan', 'Kazan'],
    'sales': [100, 250, 500, 300]
})

print(df)
```

```python
# Метод pivot_table() работает как гибрид изменения формы (.pivot()) и группировки (.groupby())

# Строим настоящую сводную таблицу
pivot_sum = pd.pivot_table(
    df,
    index='date',       # Строки (сверху вниз)
    columns='city',     # Столбцы (слева направо)
    values='sales',     # Что считаем
    aggfunc='sum',      # Как считаем (сумма)
    fill_value=0,       # Заменяем пустоты на нули
    margins=True      # Добавляем строку и столбец "Итого"

)

print(pivot_sum)
```

# Распрямление данных с помощью melt()

```python
import pandas as pd

df_wide = pd.DataFrame({
    'Company': ['Apple', 'Google'],
    '2021': [100, 150],
    '2022': [200, 250],
    '2023': [300, 350]
})

print(df_wide)
```

```python
# Плавим таблицу (столбцы превращаются в строки), оставляя Company как якорь
# variable (переменная) — сюда упадут бывшие названия столбцов (наши года).
# value (значение) — сюда упадут цифры из ячеек.

df_long = pd.melt(df_wide,
                  id_vars='Company',    # По умолчанию melt плавит все колонки, которые не указаны в id_vars
                  var_name='Year',      # Так назовем колонку с годами
                  value_name='Revenue'  # Так назовем колонку с выручкой
                  )

# Отсортируем для красоты
print(df_long.sort_values('Company'))
```

```python
# Выбор конкретных колонок для расплавления (value_vars)

df_messy = pd.DataFrame({
    'Company': ['Apple', 'Google'],
    'CEO': ['Tim Cook', 'Sundar Pichai'], # Эта колонка нам не нужна
    '2022': [200, 250],
    '2023': [300, 350]
})

# Плавим только 2022 и 2023 года, колонку CEO игнорируем
df_selected = pd.melt(
    df_messy,
    id_vars='Company',
    value_vars=['2022', '2023'],  # Явно указываем, что плавить
    var_name='Year',
    value_name='Revenue'
)

print(df_selected.head(2))
# Результат: CEO исчез, таблица аккуратно расплавилась
```

# Понятие уровней индекса

```python
# Создание мультииндекса через set_index()
import pandas as pd

# Обычная "плоская" таблица
df = pd.DataFrame({
    'Country': ['Russia', 'Russia', 'Belarus', 'Belarus'],
    'City': ['Moscow', 'Kazan', 'Minsk', 'Gomel'],
    'Sales': [1000, 500, 800, 300]
})

# Переносим ДВЕ колонки в индекс
df_multi = df.set_index(['Country', 'City'])
# Уровень 0 (level=0) — это Country (самый внешний, старший уровень).
# Уровень 1 (level=1) — это City (внутренний, младший уровень).

# мультииндекс хранится как список кортежей (tuples).

print(df_multi)
```

```python
import pandas as pd

# Создаем плоскую таблицу и сразу делаем из нее MultiIndex
df = pd.DataFrame({
    'Year': [2022, 2022, 2023, 2023],
    'Quarter': ['Q1', 'Q2', 'Q1', 'Q2'],
    'Revenue': [100, 150, 200, 250]
}).set_index(['Year', 'Quarter'])

print(df)
```

```python
# Выбор по внешнему уровню (Уровень 0)

# С самым левым (внешним) уровнем работать проще всего.
# Метод .loc[] по умолчанию обращается именно к нему
df_2022 = df.loc[2022]

print(df_2022) # теперь внешним индексом стали кварталы
```

```python
# Точечный выбор: Кортежи (Tuples)
#  Чтобы передать составной адрес для одной строки в MultiIndex, мы обязаны обернуть его в круглые скобки (кортеж).

# Передаем кортеж (Уровень_0, Уровень_1)
q1_2023 = df.loc[(2023, 'Q1')]

print(q1_2023)
```

```python
# Выбор по внутреннему уровню (Метод xs)
# Для "пробивания" индекса насквозь и поиска по любому внутреннему уровню используется  метод .xs() (от слова cross-section — поперечный срез).

# передаем в него два параметра: что ищем и на каком уровне.
# Ищем 'Q1' на уровне с именем 'Quarter' (или level=1)
df_all_q1 = df.xs('Q1', level='Quarter')

print(df_all_q1)
```

## Методы stack() и unstack()

```python
import pandas as pd

df = pd.DataFrame({
    '2022': [100, 200],
    '2023': [150, 250]
}, index=['Moscow', 'Kazan'])

df.index.name = 'City'

df
```

```python
# Этот метод берет все названия колонок и переносит их в индекс строк,
# создавая новый, самый глубокий (внутренний) уровень мультииндекса.
# Загибаем колонки (года) в строки
df_stacked = df.stack()

df_stacked
```

```python
# Метод .unstack() делает ровно обратное.
# Он берет самый внутренний уровень индекса строк (по умолчанию) и "разворачивает" его, превращая в новые колонки

# Разворачиваем самый внутренний индекс (года) обратно в колонки
df_unstacked = df_stacked.unstack()

df_unstacked
```

```python
# Но мы можем передать ему индекс любого другого уровня или его имя!
# Превращаем в колонки нулевой уровень (города)
df_city_columns = df_stacked.unstack(level='City') # аналогично level=0

df_city_columns
```

### apply()

Функция __apply()__ используется для более сложных трансформаций, в которых требуется применить функцию ко всему массиву данных. В качестве аргумента этого метода выступает функция (неважно, лямбда-функция или нет)

```python
df = pd.read_csv('https://drive.google.com/uc?export=download&id=1J1edk5yMt8tkFe7prA46mmCG5B6ORSLk')
```

```python
def age_category(age):
    if age < 30:
        return 'Young'
    elif 30 <= age < 40:
        return 'Middle-aged'
    else:
        return 'Senior'

df['Age_category'] = df['Age'].apply(age_category)
```

```python
# Тот же самый результат, но с использованием lambda-функции
df['Age_category'] = df['Age'].apply(lambda age: 'Young' if age < 30 else 'Middle-aged' if 30 <= age < 40 else 'Senior')
```

```python
# Запомните разницу:

# .apply(lambda x: ...) — вызывается для конкретного столбца дата-фрейма и работает по значениям только этого столбца;

#.apply(lambda row: ..., axis=1) — вызывается для всего дата-фрейма целиком и поэтому внутрь row передается вся строка дата-фрейма,
#благодаря чему можно обращаться к нескольким столбцам сразу. Обратите внимание что в таком случае указываем axis=1)
```

```python
# Метод .apply() позволяет применять функцию к каждому элементу в Series (столбце).

# Применим функцию, чтобы получить квадрат возраста
import pandas as pd

# Создаем DataFrame
data = {
    'Имя': ['Аня', 'Борис', 'Света'],
    'Возраст': [23, 34, 29],  # Возраст представлен строками
    'Рост': [1.65, 1.80, 1.75]  # Рост тоже строками
}

df = pd.DataFrame(data)

df['Возраст_квадрат'] = df['Возраст'].apply(lambda x: x ** 2)
print(df)
```

```python
df[['Рост']] = df[['Рост']].apply(lambda x: round(x, 1))
print(df)
```

```python
# Применение функции к каждой строке или столбцу.

# Суммируем значения в строках
df['Сумма'] = df.apply(lambda row: row['Возраст'] + row['Рост'], axis=1)  # axis=1 означает, что функция применяется к строкам
print(df)
```

## Скользящие окна

**Скользящие окна** – это инструмент, который используется для вычисления статистических метрик (таких как сумма или медиана) на основе окон данных определенной оси, которые движутся по временной оси.

В Pandas реализованы в методе **rolling()**. Его аргумент *window=* отвечает за размер окна. Рассмотрим работу функции на примере, вычислим среднее с помощью окна с размером в 5 дней:
